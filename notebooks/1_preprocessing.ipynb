```python
import polars as pl
```

```python
# Read dataset

df = pl.read_csv("../dataset/european_wholesale_electricity_price_data_daily.csv")
```

```python
# Extract the 6 different countries of:
# Austria, Germany, Switzerland, France, Poland, Estonia
countries = ["Austria", "Germany", "Switzerland", "France", "Poland", "Estonia"]
df = df.filter(pl.col("Country").is_in(countries))
```

```python
# Create a csv file with the preprocessed data for each country
for country in countries:
    df_country = df.filter(pl.col("Country") == country)
    new_df = df_country.select([
        pl.col("Date"),
        pl.col("Price (EUR/MWhe)").alias("Electricity_Price_EUR_per_MWh")
    ])
    new_df.write_csv(f"../dataset/{country.lower()}_electricity_price_daily.csv")
```

```python
# Filenames
filenames = [
    "austria_electricity_price_daily.csv",
    "germany_electricity_price_daily.csv",
    "switzerland_electricity_price_daily.csv",
    "france_electricity_price_daily.csv",
    "poland_electricity_price_daily.csv",
    "estonia_electricity_price_daily.csv"
]

print("Preprocessing completed. Generated files:")
for filename in filenames:
    print(f"- {filename}")
```
